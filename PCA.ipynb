```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette="viridis")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA on Iris Dataset")
plt.show()
```

```python
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_olivetti_faces
```

```python
# Load the Olivetti Faces dataset (images of size 64x64)
faces = fetch_olivetti_faces(shuffle=True, random_state=42)
X_faces = faces.data
```

```python
pca_faces = PCA(n_components=50)
X_faces_pca = pca_faces.fit_transform(X_faces)
```

```python
X_faces_reconstructed = pca_faces.inverse_transform(X_faces_pca)
```

```python
# Display original and reconstructed images
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for i in range(5):
    axes[0, i].imshow(X_faces[i].reshape(64, 64), cmap="gray")
    axes[0, i].set_title("Original")
    axes[1, i].imshow(X_faces_reconstructed[i].reshape(64, 64), cmap="gray")
    axes[1, i].set_title("Reconstructed")
plt.show()
```
